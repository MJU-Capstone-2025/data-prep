```python
import pandas as pd

coffee = pd.read_csv("../data/preprocessed/coffee_label.csv")
coffee
```

```python
# 'Date' 컬럼을 datetime 형식으로 변환
coffee['Date'] = pd.to_datetime(coffee['Date'])

# 연도별로 그룹화
coffee['Year'] = coffee['Date'].dt.year
annual_price = coffee.groupby('Year').agg(
    {'Coffee_Price': ['min', 'max', 'last']}).reset_index()

# 컬럼명 재정리 (정확한 순서로 컬럼명 변경)
annual_price.columns = ['Year', 'Min_Price', 'Max_Price', 'Last_Price']

# 변화율, 최대 최소 차와 비율 계산
annual_price['Price_Change'] = annual_price['Last_Price'].pct_change() + 1
annual_price['Max-Min'] = annual_price['Max_Price'] - annual_price['Min_Price']
annual_price['Max/Min'] = annual_price['Max_Price'] / annual_price['Min_Price']

# 출력
print(f"{annual_price[5:]}\n")

avg_price_change = annual_price[5:]['Price_Change'].mean()
print(f"평균 증감률: {avg_price_change}")

avg_min_max_rate = annual_price[5:]['Max/Min'].mean()
print(f"평균 증감률: {avg_min_max_rate}")
```
